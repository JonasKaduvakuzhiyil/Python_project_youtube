```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_usa = df[df['job_country'] == 'United States'].copy()
df_usa
```

```python
df_usa['job_posted_month'] = df_usa['job_posted_date'].dt.strftime('%B')
df_usa
```

```python
df_usa_pivot = df_usa.pivot_table(index = "job_posted_month", columns = 'job_title_short', aggfunc = 'size')
```

```python
df_usa_pivot.reset_index(inplace=True)
df_usa_pivot['month_number'] = pd.to_datetime(df_usa_pivot['job_posted_month'], format='%B').dt.month
df_usa_pivot.sort_values('month_number', inplace=True)
df_usa_pivot.set_index('job_posted_month', inplace=True)
df_usa_pivot.drop(columns='month_number', inplace=True)
df_usa_pivot
```

```python
df_usa_pivot.plot(kind = 'line')
```

```python
top_3 = df_usa['job_title_short'].value_counts().head(3).index
top_3 = top_3.to_list()
top_3
df_usa_pivot
```

```python
df_usa_pivot[top_3].plot(kind = 'line')
plt.title('Job Demand in the USA for Top 3 Job Titles')
plt.ylabel('Number of Job Postings')
plt.xlabel('Month')
plt.legend(title='Job Title')
plt.show()
```
