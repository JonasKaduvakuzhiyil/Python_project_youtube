```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_usa = df[df['job_country'] == 'United States'].copy()
df_usa['job_posted_month'] = df_usa['job_posted_date'].dt.strftime('%B')
df_usa_pivot = df_usa.pivot_table(index = "job_posted_month", columns = 'job_title_short', aggfunc = 'size')
df_usa_pivot.reset_index(inplace=True)
df_usa_pivot['month_number'] = pd.to_datetime(df_usa_pivot['job_posted_month'], format='%B').dt.month
df_usa_pivot.sort_values('month_number', inplace=True)
df_usa_pivot.set_index('job_posted_month', inplace=True)
df_usa_pivot.drop(columns='month_number', inplace=True)
df_usa_pivot
```

```python
df_usa_software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col= "job_posted_month")
```

```python
df_usa_merged = df_usa_pivot.merge(df_usa_software_pivot, on = 'job_posted_month')
```

```python
top_5 = (df_usa_merged
         .sum()
         .sort_values(ascending=False)
         .head(5)
         .index
         .tolist()
)
top_5
```

```python
df_usa_merged[top_5].plot(kind='line')
plt.title('Top 5 Job Titles in the USA Over Time')
plt.xlabel('Month')     
plt.ylabel('Number of Job Postings')
plt.legend(title='Job Titles', loc ='upper right')
plt.show()
```
