```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.pivot_table(values = "salary_year_avg",index="job_country",columns = "job_title_short", aggfunc = "median")
```

```python
top_countries = df['job_country'].value_counts().head(6).index
df_country_salary = df.pivot_table(values = "salary_year_avg",index="job_country",columns = "job_title_short", aggfunc = "median")
df_country_salary = df_country_salary.loc[top_countries]
job_titles = ['Data Scientist', 'Data Engineer', 'Data Analyst']

df_country_salary = df_country_salary[job_titles]
df_country_salary.plot(kind='bar')
plt.title('Average Salary by Job Title and Country')
plt.ylabel('Average Salary (Yearly)')
plt.xlabel('Country')
plt.xticks(rotation=45)
```
